```python
import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
n_groups = 5
means_frank = ( 19234+8772,806+214,6945+2267,1283+1039, 2219+842)
means_guido = (6903+1427, 35+569, 3827+1628, 735+351, 1175+408)
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, means_frank, bar_width,
alpha=opacity,
color='b',
label='verilog+perl')

 
rects2 = plt.bar(index + bar_width, means_guido, bar_width,
alpha=opacity,
color='g',
label='chisel3')
 
plt.ylabel("lines")
plt.xlabel("Tiles")
plt.legend(loc="upper right")
plt.xticks(index + bar_width, ('cdma', 'cbuf', 'csc', 'cmac', 'cacc'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

```python
countries = ['CDMA', 'CBUF', 'CSC', 'CMAC', 'CACC']
bronzes = np.array([38, 17, 26, 19, 834])
silvers = np.array([37, 23, 18, 18, 2189])
ind = [x for x, _ in enumerate(countries)]

plt.bar(ind, silvers, width=0.8, label='comments', color='silver', bottom=bronzes)
plt.bar(ind, bronzes, width=0.8, label='codes', color='#CD853F')

plt.xticks(ind, countries)
plt.ylabel("lines")
plt.xlabel("Tiles")
plt.legend(loc="upper right")
plt.title("Lines of Code Statistic Result")

plt.show()
```
